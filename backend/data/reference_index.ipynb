```python
import pandas as pd
import yfinance as yf
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib

# 맑은 고딕 폰트 설정
matplotlib.rc('font', family='Malgun Gothic')
matplotlib.rcParams['axes.unicode_minus'] = False  # 마이너스 기호 깨짐 방지

# 1. 유효한 티커들 (필터링된 버전 사용)
tickers = [
    'ACWI', 'ASEA', 'URTH', 'EMB', 'INDA', 'MCHI', '^KS200', 'VPL', '^GSPC', 'NA.TO',
    'AGG', '000300.SS', 'VNM', '^HSCE', '^NDX', '^KS11', '^N225', 'IBB'
]

start_date = "2023-07-01"
end_date = "2025-07-01"

adj_close_dict = {}
failures = []

for t in tickers:
    try:
        ticker_df = yf.download(t, start=start_date, end=end_date, progress=False)
        if isinstance(ticker_df, pd.DataFrame) and not ticker_df.empty:
            # 'Adj Close'가 없으면 'Close'를 사용
            if 'Adj Close' in ticker_df.columns:
                series = ticker_df['Adj Close'].dropna()
            elif 'Close' in ticker_df.columns:
                series = ticker_df['Close'].dropna()
            else:
                raise ValueError("No valid price column")
            if not series.empty:
                adj_close_dict[t] = series
                print(f"✅ {t} 수집 완료")
            else:
                print(f"⚠️ {t}: 시계열 없음")
                failures.append(t)
        else:
            print(f"⚠️ {t}: DataFrame 비어있음")
            failures.append(t)
    except Exception as e:
        print(f"❌ {t} 실패: {e}")
        failures.append(t)

# 2. 수익률 계산 및 시각화
if adj_close_dict:
    df_prices = pd.concat(adj_close_dict, axis=1)
    df_returns = df_prices.pct_change().dropna()
    df_cum = (1 + df_returns).cumprod() - 1

    # 시각화
    plt.figure(figsize=(14, 7))
    df_cum.plot()

    plt.title("지수별 누적 수익률 (1년)")
    plt.xlabel("날짜")
    plt.ylabel("누적 수익률")
    plt.grid(True)

    # 범례 이름을 티커로 설정
    plt.legend(title='Ticker', bbox_to_anchor=(1.05, 1), loc='upper left')

    # 그래프 밖으로 범례 배치
    plt.tight_layout()
    plt.show()
else:
    print("⚠️ 수익률 계산에 사용할 데이터가 없습니다.")

print("\n❌ 실패한 티커:", failures)
```

```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
import matplotlib

# 한글 폰트 설정
matplotlib.rc('font', family='Malgun Gothic')
matplotlib.rcParams['axes.unicode_minus'] = False

def auto_map_ticker(name):
    if pd.isna(name): return None
    name = name.upper()
    if "ACWI" in name: return "ACWI"
    if "INFRA" in name: return "IFRA"
    if "INDIA" in name: return "INDA"
    if "NORTH AMERICA" in name: return "NA.TO"
    if "WORLD" in name or "WI" in name: return "URTH"
    if "CHINA" in name: return "MCHI"
    if "SOUTH EAST ASIA" in name: return "ASEA"
    if "EMERGING" in name: return "EEM"
    if "PACIFIC" in name: return "VPL"
    if "AAXJ" in name: return "AAXJ"
    if "NASDAQ 100" in name or "나스닥 100" in name: return "^NDX"
    if "NASDAQ" in name: return "^IXIC"
    if "S&P" in name: return "^GSPC"
    if "DOW JONES" in name: return "^DJI"
    if "EMBI" in name: return "EMB"
    if "AGGREGATE" in name or "BARCLAYS" in name: return "AGG"
    if "CD" in name or "국공채" in name or "종합채권" in name: return None
    if "NIKKEI" in name: return "^N225"
    if "CSI 300" in name: return "000300.SS"
    if "HSCE" in name or "항셍" in name: return "^HSCE"
    if "VNM" in name or "베트남" in name: return "VNM"
    if "KOSPI200" in name: return "^KS200"
    if "KOSPI" in name: return "^KS11"
    if "코스닥 150" in name: return None
    if "DRIV" in name or "ELECTRIC" in name: return "DRIV"
    if "IBB" in name or "생명과학" in name: return "IBB"
    return None

# 데이터 로딩 및 컬럼 정리
df = pd.read_csv('new_irp_non.csv', encoding='utf-8')
df = df.drop(['Unnamed: 0.1', 'Unnamed: 0'], axis=1, errors='ignore')

# 매핑지수 컬럼 생성 및 타입 안전 처리
df['매핑지수'] = df['지수1'].apply(auto_map_ticker)
df['매핑지수'] = df['매핑지수'].where(pd.notna(df['매핑지수']), None)
df['매핑지수'] = df['매핑지수'].apply(lambda x: x if isinstance(x, str) else None)

# 수익률 계산 함수 (float만 반환)
def get_index_return(index_ticker, start_date, end_date):
    if not index_ticker or not isinstance(index_ticker, str):
        return 0.0
    try:
        data = yf.download(index_ticker, start=start_date, end=end_date, progress=False)
        if 'Adj Close' in data.columns:
            adj_close = data['Adj Close'].dropna()
        elif 'Close' in data.columns:
            adj_close = data['Close'].dropna()
        else:
            print(f"❌ {index_ticker}: 'Adj Close' 또는 'Close' 컬럼이 없습니다.")
            return 0.0
        if len(adj_close) < 2:
            return 0.0
        return float((adj_close.iloc[-1] / adj_close.iloc[0]) - 1)  # float로 명시적 변환
    except Exception as e:
        print(f"❌ {index_ticker} 다운로드 실패: {e}")
        return 0.0

# 날짜 범위 설정
start_date = "2020-07-01"
end_date = "2025-06-30"

# for문으로 수익률 계산
result_list = []
for idx, row in df.iterrows():
    ticker = row['매핑지수']
    if isinstance(ticker, str):
        ret = get_index_return(ticker, start_date, end_date)
    else:
        ret = 0.0
    result_list.append(ret)
df['지수수익률'] = result_list

# 결과 출력
df.to_csv('update_irp_non.csv')

```
